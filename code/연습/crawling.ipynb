```python
from newspaper import Article
import re 
from summa.summarizer import summarize
def crawling():
    url = input('url을 입력하세요')
    article = Article(url,laguage='ko')
    article.download()
    article.parse()
    title = article.title
    text = article.text
    text = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s,.%()]", "", text)
    text_sum = summarize(text, ratio=0.2)
    return title , text , text_sum
```

```python
def crawling(url):
    # url = input('url을 입력하세요')
    article = Article(url,laguage='ko')
    article.download()
    article.parse()
    title = article.title
    text = article.text
    text = '.'.join(article.text.split('.')[:-2:])
    text = ''.join(text.split('기자) ')[1]) # split 할 경우 
    text = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s,.%()]", "", text)
    text_sum = '\n'.join(summarize(text).split('.'))
    return title , text , print(text_sum)
```
